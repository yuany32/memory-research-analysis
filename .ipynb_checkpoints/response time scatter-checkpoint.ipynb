```python
#the list of training & testing fnames
fnames = sorted(glob.glob('*.csv'))
training_files = [fname for fname in fnames if re.match('(\d*)_main_exp_(\w*)_(\w*)_(\w*)_(\d*_\w*_\d*)_(\d*)_data.csv', fname)]
testing_files = [fname for fname in fnames if re.match('(\d*)_main_exp_(\w*)_(\w*)_(\w*)_(\d*_\w*_\d*)_(\d*).csv', fname)]
```

```python
#adding file_name colomn to each file
list_of_dfs = [pd.read_csv(fname) for fname in training_files]
for dataframe, fname in zip(list_of_dfs, training_files):
    dataframe['file_name'] = fname
```

```python
# Combine a list of dataframes, on top of each other
training = pd.concat(list_of_dfs, ignore_index=True)
```

```python
#merging block_order, block_order_bytype to each training file
info = pd.read_csv("fnames_info.csv")
training_data = pd.merge(training,
                 info[['file_name', 'block_order', 'block_order_bytype','subj_id']],
                 on ='file_name')
training_data = training_data.sort_values(by = ['subj_id','block_order'])
training_data = training_data.reset_index(drop = True)
training_data.head()
```

```python
# get data for d1 and d2
def dpoint_rt(subj_id,block,criteria):
    # select specific subject and block data from trainig_data
    subj_acc = training_data.loc[training_data.subj_id == subj_id]
    block_data = subj_acc.loc[subj_acc.block_order == block]
    block_data = block_data.reset_index(drop = True)
    
    # select decision 1 and decision 2 rows into two different df
    d1_data = block_data.loc[block_data.pic_type == "decision1"].reset_index(drop = True)
    d2_data = block_data.loc[block_data.pic_type == "decision2"].reset_index(drop = True)
    
    # Decision point 1 rt
    d1_rt = d1_data['rt']
    d2_rt = d2_data['rt']
    
    # Decision point 2 rt
    
    d1_data['d1_rt'] = d1_rt
    d2_data['d2_rt'] = d2_rt
    
    
    d1_rt_sum = []
    d2_rt_sum = []
    d1_correct_rt_sum = []
    d1_incorrect_rt_sum = []
    d2_correct_rt_sum = []
    d2_incorrect_rt_sum = []
    
    for i in range (0,len(d1_data)-(criteria-1)):
        iterate_dpoint_1 = ((d1_data.iloc[i:(i+criteria)])['d1_rt']).sum()/criteria
        iterate_dpoint_2 = ((d2_data.iloc[i:(i+criteria)])['d2_rt']).sum()/criteria
        d1_rt_sum.append(iterate_dpoint_1)
        d2_rt_sum.append(iterate_dpoint_2)
        iterate_correct_dpoint_1 = ((d1_data.iloc[i:(i+criteria)]))
        iterate_correct = iterate_correct_dpoint_1.loc[iterate_correct_dpoint_1.feedback_code == 1]['rt'].sum()/criteria
        iterate_incorrect = iterate_correct_dpoint_1.loc[iterate_correct_dpoint_1.feedback_code != 1]['rt'].sum()/criteria
        d1_correct_rt_sum.append(iterate_correct)
        d1_incorrect_rt_sum.append(iterate_incorrect)
        iterate_correct_dpoint_2 = ((d2_data.iloc[i:(i+criteria)]))
        iterate_correct = iterate_correct_dpoint_2.loc[iterate_correct_dpoint_2.feedback_code == 1]['rt'].sum()/criteria
        iterate_incorrect = iterate_correct_dpoint_2.loc[iterate_correct_dpoint_2.feedback_code != 1]['rt'].sum()/criteria
        d2_correct_rt_sum.append(iterate_correct)
        d2_incorrect_rt_sum.append(iterate_incorrect) 
        
    def zero_to_nan(values):
    #"""Replace every 0 with 'nan' and return a copy."""
        return [float('nan') if x==0 else x for x in values]    
    #create new dataframe for all data
    dpoint_rt = pd.DataFrame(columns = ['trial_criteria','d1_rt_sum','d2_rt_sum',
                                        'd1_correct','d1_incorrect','d2_correct','d2_incorrect'])
    dpoint_rt['trial_criteria'] = np.arange(len(d1_data)-(criteria-1)) + 1
    dpoint_rt['d1_rt_sum'] = d1_rt_sum
    dpoint_rt['d2_rt_sum'] = d2_rt_sum
    dpoint_rt['d1_correct'] = zero_to_nan(d1_correct_rt_sum)
    dpoint_rt['d1_incorrect'] = zero_to_nan(d1_incorrect_rt_sum)
    dpoint_rt['d2_correct'] = zero_to_nan(d2_correct_rt_sum)
    dpoint_rt['d2_incorrect'] = zero_to_nan(d2_incorrect_rt_sum)
    dpoint_rt['trial_target'] = block_data['trial_target'][0] 
    
    return dpoint_rt
```

```python
def path_rt (subj_id,block,criteria):
    # select specific subject and block data from trainig_data
    subj_acc = training_data.loc[training_data.subj_id == subj_id]
    block_data = subj_acc.loc[subj_acc.block_order == block]
    block_data = block_data.reset_index(drop = True)
    
    # select path1,2,3 rows
    path_data = block_data.loc[block_data.segment == "path"].reset_index(drop = True)
    
    path_rt = []
    path_correct_rt = []
    path_incorrect_rt = []
    for i in range (1,path_data['trial_num'].iloc[-1]+1): # <------- use trial_num as a indicator for iterating through rows
        rt_sum = path_data.loc[path_data['trial_num'] == i, 'rt'].sum()/3
        path_rt.append(rt_sum)
        i += 1 
    feedback_code = block_data.loc[block_data.segment == 'target'].reset_index(drop = True)['feedback_code']
    path_rt_sum = pd.DataFrame(columns=['trial_criteria','path_rt','feedback_code'])  
    path_rt_sum['trial_criteria'] = np.arange(len(path_rt))+1
    path_rt_sum['path_rt'] = path_rt
    path_rt_sum['feedback_code'] = feedback_code
    rt_sum_criteria = []
    for i in range(0,len(path_rt) - (criteria-1)):
        iterate_path = (path_rt_sum.iloc[i:(i+criteria)])['path_rt'].sum()/criteria
        rt_sum_criteria.append(iterate_path)
        iterate_correct_path = path_rt_sum.iloc[i:(i+criteria)]
        iterate_correct = (iterate_correct_path.loc[iterate_correct_path.feedback_code == 1])['path_rt'].sum()/criteria
        iterate_incorrect = (iterate_correct_path.loc[iterate_correct_path.feedback_code != 1])['path_rt'].sum()/criteria
        path_correct_rt.append(iterate_correct)
        path_incorrect_rt.append(iterate_incorrect)
    def zero_to_nan(values):
    #"""Replace every 0 with 'nan' and return a copy."""
        return [float('nan') if x==0 else x for x in values]  
    
    path_rt_sum = pd.DataFrame(columns=['trial_criteria','rt_sum_criteria','correct_path','incorrect_path','trial_target']) 
    path_rt_sum['trial_criteria'] = np.arange(len(rt_sum_criteria))+1
    path_rt_sum['rt_sum_criteria'] = rt_sum_criteria
    path_rt_sum['correct_path'] = zero_to_nan(path_correct_rt)
    path_rt_sum['incorrect_path'] = zero_to_nan(path_incorrect_rt)
    path_rt_sum['trial_target'] =  block_data['trial_target'][0]

    return path_rt_sum   
```

```python
# get data for d1 and d2 look at each target individually
def dpoint_criteria(subj_id,block,criteria):
    # select specific subject and block data from trainig_data
    subj_acc = training_data.loc[training_data.subj_id == subj_id]
    block_data = subj_acc.loc[subj_acc.block_order == block]
    block_data = block_data.reset_index(drop = True)
    
    # select decision 1 and decision 2 rows into two different df
    d1_data = block_data.loc[block_data.pic_type == "decision1"].reset_index(drop = True)
    d2_data = block_data.loc[block_data.pic_type == "decision2"].reset_index(drop = True)
    d1_correct = []
    d2_correct = []
    
    # Decision point 1 correctness
    for i in range(0,len(d1_data)):
        if (d1_data['trial_target'][i] == "lion"):
            if(d1_data['pic_name'][i] == "goose"):
                d1_correct.append(True)
            else:
                d1_correct.append(False)
        elif (d1_data['trial_target'][i] == "bolt"):
            if (d1_data['pic_name'][i] == "domino"):
                d1_correct.append(True)
            else:
                d1_correct.append(False)
        elif (d1_data['trial_target'][i] == "beaver"):
            if(d1_data['pic_name'][i] == "pigeon"):
                d1_correct.append(True)
            else:
                d1_correct.append(False)
        elif (d1_data['trial_target'][i] == "arc"):
            if(d1_data['pic_name'][i] == "watch"):
                d1_correct.append(True)
            else:
                d1_correct.append(False)
    
    # Decision point 2 correctness
    for i in range(0,len(d2_data)):
        if (d2_data['trial_target'][i] == "lion"):
            if(d2_data['pic_name'][i] == "crutches"):
                d2_correct.append(True)
            else:
                d2_correct.append(False)
        elif (d2_data['trial_target'][i] == "bolt"):
            if (d2_data['pic_name'][i] == "garlic"):
                d2_correct.append(True)
            else:
                d2_correct.append(False)
        elif (d2_data['trial_target'][i] == "beaver"):
            if(d2_data['pic_name'][i] == "stepladder"):
                d2_correct.append(True)
            else:
                d2_correct.append(False)
        elif (d2_data['trial_target'][i] == "arc"):
            if(d2_data['pic_name'][i] == "holly"):
                d2_correct.append(True)
            else:
                d2_correct.append(False)
    d1_data['d1_correct'] = (d1_correct)  
    d2_data['d2_correct'] = (d2_correct)  
    
    #count the correct times per criteria
    d1_count = 0 
    d1_percent = 0
    d1_correct_count = []
    d1_correct_percent = []
    d2_count = 0 
    d2_percent = 0
    d2_correct_count = []
    d2_correct_percent = []
    for i in range (0,len(d1_data)-(criteria-1)):
        iterate_dpoint_1 = (d1_data.iloc[i:(i+criteria)])['d1_correct']
        iterate_dpoint_2 = (d2_data.iloc[i:(i+criteria)])['d2_correct']
        for iterate_rows in iterate_dpoint_1:
            if iterate_rows:
                d1_count += 1
                d1_percent = d1_count / criteria
        d1_correct_count.append(d1_count)
        d1_correct_percent.append(d1_percent)
        d1_count = 0
        d1_percent = 0
        for iterate_rows in iterate_dpoint_2:
            if iterate_rows:
                d2_count += 1
                d2_percent = d2_count / criteria
        d2_correct_count.append(d2_count)
        d2_correct_percent.append(d2_percent)
        d2_count = 0
        d2_percent = 0
        
    #create new dataframe for all data
    dpoint_criteria = pd.DataFrame(columns = ['trial_criteria',
                                              'd1_percent','d1_count','d2_percent','d2_count'])
    dpoint_criteria['trial_criteria'] = np.arange(len(d1_data)-(criteria-1)) + 1
    dpoint_criteria['d1_percent'] = d1_correct_percent
    dpoint_criteria['d1_count'] = d1_correct_count
    dpoint_criteria['d2_percent'] = d2_correct_percent
    dpoint_criteria['d2_count'] = d2_correct_count
    dpoint_criteria['trial_target'] = block_data['trial_target'][0] 
    dpoint_criteria['trial_sq'] = dpoint_criteria['trial_criteria']**2
    
    return dpoint_criteria
```

```python
# get data for d1 and d2 ignore specific target
def overall_dpoint_criteria(subj_id,block,criteria):
    # select specific subject and block data from trainig_data
    subj_acc = training_data.loc[training_data.subj_id == subj_id]
    block_data = subj_acc.loc[subj_acc.block_order == block]
    block_data = block_data.reset_index(drop = True)
    
    # select decision 1 and decision 2 rows into two different df
    d1_data = block_data.loc[block_data.pic_type == "decision1"].reset_index(drop = True)
    d2_data = block_data.loc[block_data.pic_type == "decision2"].reset_index(drop = True)
    d1_correct = []
    d2_correct = []
    
    # Decision point 1 correctness
    for i in range(0,len(d1_data)):
        if (d1_data['feedback_code'][i] == 1):
            d1_correct.append(True)
        else:
            d1_correct.append(False)
    
    # Decision point 2 correctness
    for i in range(0,len(d2_data)):
        if (d2_data['feedback_code'][i] == 1):
            d2_correct.append(True)
        else:
            d2_correct.append(False)
            
    d1_data['d1_correct'] = (d1_correct)  
    d2_data['d2_correct'] = (d2_correct)  
    
    #count the correct times per criteria
    d1_count = 0 
    d1_percent = 0
    d1_correct_count = []
    d1_correct_percent = []
    d2_count = 0 
    d2_percent = 0
    d2_correct_count = []
    d2_correct_percent = []
    for i in range (0,len(d1_data)-(criteria-1)):
        iterate_dpoint_1 = (d1_data.iloc[i:(i+criteria)])['d1_correct']
        iterate_dpoint_2 = (d2_data.iloc[i:(i+criteria)])['d2_correct']
        for iterate_rows in iterate_dpoint_1:
            if iterate_rows:
                d1_count += 1
                d1_percent = d1_count / criteria
        d1_correct_count.append(d1_count)
        d1_correct_percent.append(d1_percent)
        d1_count = 0
        d1_percent = 0
        for iterate_rows in iterate_dpoint_2:
            if iterate_rows:
                d2_count += 1
                d2_percent = d2_count / criteria
        d2_correct_count.append(d2_count)
        d2_correct_percent.append(d2_percent)
        d2_count = 0
        d2_percent = 0
        
    #create new dataframe for all data
    dpoint_criteria = pd.DataFrame(columns = ['trial_criteria',
                                              'd1_percent','d1_count','d2_percent','d2_count'])
    dpoint_criteria['trial_criteria'] = np.arange(len(d1_data)-(criteria-1)) + 1
    dpoint_criteria['d1_percent'] = d1_correct_percent
    dpoint_criteria['d1_count'] = d1_correct_count
    dpoint_criteria['d2_percent'] = d2_correct_percent
    dpoint_criteria['d2_count'] = d2_correct_count
    dpoint_criteria['trial_target'] = block_data['trial_target'][0] 
    dpoint_criteria['trial_sq'] = dpoint_criteria['trial_criteria']**2
    
    return dpoint_criteria
```

```python
def max_rt_correct(subj_id,criteria):
    max_rt = 0
    for j in range(1,5):
        rt_criteria = dpoint_correct_rt(subj_id,j,criteria)
        path_rt_sum = path_rt (subj_id = subj_id,block = j,criteria = criteria)
        arm1_rt_sum = arm1_rt (subj_id = subj_id,block = j,criteria = criteria)
        arm2_rt_sum = arm2_rt (subj_id = subj_id,block = j,criteria = criteria)
        
        max_rt_d1_correct = rt_criteria['d1_correct'].max()
        max_rt_d1_incorrect = rt_criteria['d1_incorrect'].max()
        max_rt_d2_correct = rt_criteria['d2_correct'].max()
        max_rt_d2_incorrect = rt_criteria['d2_incorrect'].max() 
        max_rt_p = path_rt_sum['rt_sum_criteria'].max()
        max_rt_arm1 = arm1_rt_sum['rt_sum_criteria'].max()
        max_rt_arm2 = arm2_rt_sum['rt_sum_criteria'].max()
        max_list = [max_rt_d1_correct,max_rt_d1_incorrect,
                    max_rt_d2_correct,max_rt_d2_correct,
                    max_rt_p,max_rt_arm1,
                    max_rt_arm2]
        max_num = max(max_list)
        if max_num > max_rt:
            max_rt = max_num
    return max_rt
```

```python
def max_rt_overall(subj_id,criteria):
    max_rt = 0
    for j in range(1,5):
        rt_criteria = dpoint_rt(subj_id,j,criteria)
        path_rt_sum = path_rt (subj_id = subj_id,block = j,criteria = criteria)
        arm1_rt_sum = arm1_rt (subj_id = subj_id,block = j,criteria = criteria)
        arm2_rt_sum = arm2_rt (subj_id = subj_id,block = j,criteria = criteria)
        
        max_rt_d1 = rt_criteria['d1_rt_sum'].max()
        max_rt_d2 = rt_criteria['d2_rt_sum'].max() 
        max_rt_p = path_rt_sum['rt_sum_criteria'].max()
        max_rt_arm1 = arm1_rt_sum['rt_sum_criteria'].max()
        max_rt_arm2 = arm2_rt_sum['rt_sum_criteria'].max()
        max_list = [max_rt_d1,max_rt_d2,max_rt_p,max_rt_arm1,
                    max_rt_arm2]
        max_num = max(max_list)
        if max_num > max_rt:
            max_rt = max_num
    return max_rt
```

```python
# graph decision point criteria & rt for all points without separating correctness
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit 
import scipy.optimize as optimize
import matplotlib.patches as mpatches
def rt_graph(subj_id,criteria):
    # learning curve 
    fig, axes = plt.subplots(nrows=5, ncols=4, figsize=(20,25))
    
    for j in range(1,5):
        rt_criteria = dpoint_rt(subj_id,j,criteria)
        path_rt_sum = path_rt(subj_id = subj_id,block = j,criteria = criteria)
        arm1_rt_sum = arm1_rt(subj_id = subj_id,block = j,criteria = criteria)
        arm2_rt_sum = arm2_rt(subj_id = subj_id,block = j,criteria = criteria)
        dp_criteria = dpoint_criteria(subj_id = subj_id,block = j,criteria = criteria)
         # scatter plot of learning criteria
        target = rt_criteria['trial_target'][0]
        if target == "arc":
            target = "beaver"
        elif target == "bolt":
            target = "lion"
        
        #path log formula
        def func(t, a, b):
            return a + b * np.log(t)
        popt, pcov = optimize.curve_fit(func,path_rt_sum['trial_criteria'] , path_rt_sum['rt_sum_criteria'], maxfev=5000)
        t = rt_criteria['trial_criteria']
        
        #path rt scatter
        plt.tight_layout()
        axes[0,j-1].set_title('path rt (criteria = ' + str(criteria) + ')',loc = 'center',fontsize=14)
        axes[0,j-1].scatter(path_rt_sum['trial_criteria'],path_rt_sum['rt_sum_criteria'],label = 'subj_' + str(subj_id) \
                + ', block ' + str(j) + ', ' + target + ', rt' ,color = 'red')
        axes[0,j-1].plot(path_rt_sum['trial_criteria'], func(path_rt_sum['trial_criteria'],*popt), label = '',
                       linewidth = 3,color = 'red') # <------------- Add predicted values  
        axes[0,j-1].legend(loc = "best")
        
        #d1 log formula
        def func(t, a, b):
            return a + b * np.log(t)
        popt, pcov = optimize.curve_fit(func,rt_criteria['trial_criteria'] , rt_criteria['d1_rt_sum'], maxfev=5000)
        t = rt_criteria['trial_criteria']
        
        #d1 rt scatter
        plt.tight_layout()
        axes[1,j-1].set_title('dpoint1 rt (criteria = ' + str(criteria) + ')',loc = 'center',fontsize=14)
        axes[1,j-1].scatter(rt_criteria['trial_criteria'],rt_criteria['d1_rt_sum'],label = 'subj_' + str(subj_id) \
                + ', block ' + str(j) + ', ' + target + ', rt' ,color = 'green')
        axes[1,j-1].plot(rt_criteria['trial_criteria'], func(rt_criteria['trial_criteria'],*popt), label = '',
                      linewidth = 3,color = 'green') # <------------- Add predicted values  
        axes[1,j-1].legend(loc = 'best')
        
        #arm1 log formula
        def func(t, a, b):
            return a + b * np.log(t)
        popt, pcov = optimize.curve_fit(func,arm1_rt_sum['trial_criteria'] , arm1_rt_sum['rt_sum_criteria'], maxfev=5000)
        t = rt_criteria['trial_criteria']
        
        #arm1 rt scatter
        plt.tight_layout()
        axes[2,j-1].set_title('arm1 rt (criteria = ' + str(criteria) + ')',loc = 'center',fontsize=14)
        axes[2,j-1].scatter(arm1_rt_sum['trial_criteria'],arm1_rt_sum['rt_sum_criteria'],label = 'subj_' + str(subj_id) \
                + ', block ' + str(j) + ', ' + target + ', rt' ,color = 'blue')
        axes[2,j-1].plot(arm1_rt_sum['trial_criteria'], func(arm1_rt_sum['trial_criteria'],*popt), label = '',
                       linewidth = 3,color = 'blue') # <------------- Add predicted values  
        axes[2,j-1].legend(loc = "best")
        
        #d2 log formula
        def func(t, a, b):
            return a + b * np.log(t)
        popt, pcov = optimize.curve_fit(func,rt_criteria['trial_criteria'] , rt_criteria['d2_rt_sum'], maxfev=5000)
        t = rt_criteria['trial_criteria']
        
        #d2 rt scatter
        axes[3,j-1].set_title('dpoint2 rt (criteria = ' + str(criteria) + ')',loc = 'center',fontsize=14)
        axes[3,j-1].scatter(rt_criteria['trial_criteria'],rt_criteria['d2_rt_sum'],label = 'subj_' + str(subj_id) \
                + ', block ' + str(j) + ', ' + target + ', rt' ,color = 'orange')
        axes[3,j-1].plot(rt_criteria['trial_criteria'], func(rt_criteria['trial_criteria'],*popt), label = '',
                       linewidth = 3,color = 'orange') # <------------- Add predicted values
        axes[3,j-1].legend(loc='best')
        
        #arm2 log formula
        def func(t, a, b):
            return a + b * np.log(t)
        popt, pcov = optimize.curve_fit(func,arm2_rt_sum['trial_criteria'] , arm2_rt_sum['rt_sum_criteria'], maxfev=5000)
        t = rt_criteria['trial_criteria']
        
        # arm2 rt scatter
        axes[4,j-1].set_title('arm2 rt (criteria = ' + str(criteria) + ')',loc = 'center',fontsize=14)
        axes[4,j-1].scatter(arm2_rt_sum['trial_criteria'],arm2_rt_sum['rt_sum_criteria'],label = 'subj_' + str(subj_id) \
                + ', block ' + str(j) + ', ' + target + ', rt' ,color = 'purple')
        axes[4,j-1].plot(arm2_rt_sum['trial_criteria'], func(arm2_rt_sum['trial_criteria'],*popt), label = '',
                       linewidth = 3,color = 'purple') # <------------- Add predicted values  
        axes[4,j-1].legend(loc = "upper right")
        
        # scatter plot settings
        max_rt = max_rt_overall(subj_id = subj_id, criteria = criteria)
        
        for i in range(0,5):
            axes[i,j-1].set_xlabel('trial bins',fontsize=14) # <--------------- Set x axis label
            axes[i,j-1].set_yticks(np.arange(0,(max_rt + 1),step=(max_rt / 5)))
            axes[i,j-1].set_ylim(0, max_rt)
            axes[i,j-1].set_xlim(0,len(rt_criteria['trial_criteria']))
            axes[i,j-1].set_ylabel('response time (average)',fontsize=14) # <--------- Set y axis label    
        plt.rc('xtick',labelsize= 14)
        plt.rc('ytick',labelsize= 14)
        
        j += 1   
        
    #save plots
    
    #plt.savefig('subj'+str(subj_id)+ '_criteria_'+ str(criteria)+'_rt.png')
    #plt.close() 
```
