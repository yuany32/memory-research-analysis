```python
def trial_rt (subj_id,block):
    # select specific subject and block data from trainig_data
    subj_acc = training_data.loc[training_data.subj_id == subj_id]
    block_data = subj_acc.loc[subj_acc.block_order == block]
    block_data = block_data.reset_index(drop = True)
    feedback_sum = []
    for i in range (1,block_data['trial_num'].iloc[-1]+1): # <------- use trial_num as a indicator for iterating through rows
        trial_sum = (block_data.loc[block_data['trial_num'] == i, 'rt'].sum())
        feedback_sum.append(trial_sum)
    block_rt = pd.DataFrame(columns = ['trial_num'])
    block_rt['trial_num'] = np.arange(len(feedback_sum)) + 1
    block_rt['feedback_sum'] = feedback_sum
    block_rt['trial_target'] = block_data['trial_target'][0] 
    return block_rt
print(trial_rt(1,1))
```

```python
def mask_ok_table(subj_id,outlier):
    subj_mask_table = pd.DataFrame(columns = ["subj_id","block_order","trial_num","trial_rt_sum","mask_ok","trial_target"])
    subj_mask = []
    trial_num = []
    block_order = []
    target_list = []
    subj_acc = training_data.loc[training_data.subj_id == subj_id]
    for block in range(1,5):
        block_data = subj_acc.loc[subj_acc.block_order == block]
        block_data = block_data.reset_index(drop = True)
        trial = list(range(1,block_data['trial_num'].iloc[-1]+1))
        blocks = list(str(block)*len(trial))
        target = [block_data['trial_target'][0]]
        feedback_sum = []
        trial_num += trial
        block_order += blocks
        targets = [item for item in target for i in range(len(trial))]
        target_list += targets
        for i in range (1,block_data['trial_num'].iloc[-1]+1): # <------- use trial_num as a indicator for iterating through rows
            trial_sum = (block_data.loc[block_data['trial_num'] == i, 'rt'].sum())
            feedback_sum.append(trial_sum)
        subj_mask+=(feedback_sum)
    #stdev = subj_mask_ok['all_rt'].std()
    subj_mask_table['trial_target'] = target_list
    subj_mask_table['block_order'] = block_order
    subj_mask_table['trial_num'] = trial_num
    subj_mask_table['trial_rt_sum'] = subj_mask
    stdev = subj_mask_table['trial_rt_sum'].std()
    mean = subj_mask_table['trial_rt_sum'].mean()
    outlier = mean + stdev * outlier
    subj_mask_table['subj_id'] = [str(subj_id)] * len(target_list)
    #print("std = " + str(stdev) + ", outlier range = " + str(outlier))
    subj_mask_table.loc[subj_mask_table['trial_rt_sum'] <= outlier, 'mask_ok'] = True
    subj_mask_table.loc[subj_mask_table['trial_rt_sum'] > outlier, 'mask_ok'] = False
    return(subj_mask_table)
      
mask_ok_table(2,3)
```

```python
def group_mask_ok_table(outlier):
    #group_mask = pd.DataFrame(columns = ["subj_id","block_order","trial_num","trial_rt_sum","mask_ok","trial_target"])
    group_mask = []
    for subj in range(1,31):
        df = mask_ok_table(subj_id = subj,outlier = 3)
        group_mask.append(df)
    group_mask = pd.concat(group_mask)
    return(group_mask)
      
group_mask_ok_table(3)
```

```python
def trial_rt_graph(subj_id,outlier):
    fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 5))
    rt = mask_ok_table(subj_id,outlier)
    max_sum = rt['trial_rt_sum'].max()
    for j in range(1,5):
        rt_sum = rt.loc[rt.block_order == str(j)].reset_index(drop=True)
        max_sum = rt['trial_rt_sum'].max().round()
        # scatter plot of learning criteria
        target = rt_sum['trial_target'][0]
        if target == "arc":
            target = "beaver"
        elif target == "bolt":
            target = "lion"
        trial_ok = rt_sum.loc[rt_sum.mask_ok == True,'trial_num'].reset_index(drop=True)
        trial_failed = rt_sum.loc[rt_sum.mask_ok == False,'trial_num'].reset_index(drop=True)
        mask_ok = rt_sum.loc[rt_sum.mask_ok == True, 'trial_rt_sum'].reset_index(drop=True)
        mask_failed = rt_sum.loc[rt_sum.mask_ok == False, 'trial_rt_sum'].reset_index(drop=True)
        plt.tight_layout()
        axes[j-1].set_title('trial rt mask',loc = 'center')
        axes[j-1].scatter(trial_ok,mask_ok,label = 'subj_' + str(subj_id) \
                + ', block ' + str(j) + ', ' + target,color = "green" )
        axes[j-1].scatter(trial_failed,mask_failed,label = 'subj_' + str(subj_id) \
                + ', block ' + str(j) + ', failed mask, ' + target,marker = 'x',color = "red" )
        axes[j-1].set_ylim(0, max_sum)
        axes[j-1].tick_params(axis="y",labelsize=12, labelrotation=0, labelcolor="black")
        axes[j-1].set_xlabel('trial bins') # <--------------- Set x axis label
        axes[j-1].legend(loc = "upper right")
        #axes[0,j-1].set_xlim(0, len(rt_sum['trial_num']))
        axes[j-1].set_ylabel('rt (trial sum)') 
    #plt.rc('xtick',labelsize= 16)
    #plt.rc('ytick',labelsize= 16)

    #save subplots for each subject
    plt.savefig('subj'+str(subj_id)+ '_mask_ok_' + str(outlier) + '.png')
    plt.close() 
#trial_rt_graph(subj_id = 12,outlier = 3)
```

```python
def save_graph(subj_id,outlier):
    for i in range(1,subj_id+1):
        trial_rt_graph(i,outlier) 
save_graph(30,3)
```

```python
def block_rt_sum(subj_id, block, bins):
    subj_data = training_data.loc[(training_data.subj_id == subj_id).reset_index(drop = True)]
    block_data = subj_data.loc[(subj_data.block_order == block)].reset_index(drop=True)
    d2_rt = block_data.loc[block_data.pic_type == "decision2"].reset_index(drop=True)
    arm1_rt = block_data.loc[block_data.pic_type == "arm21"].reset_index(drop=True)
    arm2_rt = block_data.loc[block_data.pic_type == "arm22"].reset_index(drop=True)
    arm3_rt = block_data.loc[block_data.pic_type == "arm23"].reset_index(drop=True)
    block_rt = pd.DataFrame(columns = ['block_order','trial_end','trial_bin','d2_rt','arm21_rt','arm22_rt','arm23_rt'])
    d2 = []
    arm21 = []
    arm22 = []
    arm23 = []
    block = []
    d2_err = []
    arm21_err = []
    arm22_err = []
    arm23_err = []
    bins_scalar = []
    mult_number = int(round((block_data['trial_num'].iloc[-1])/bins))+1
    i_series = range(0,mult_number)
    scale = 1
    for scalar in i_series:
        bin_end = scalar * bins
        bins_scalar.append(bin_end)
    d2_rt['rt_bins'] = pd.cut(x=d2_rt['trial_num'], bins=bins_scalar)
    arm1_rt['rt_bins'] = pd.cut(x=arm1_rt['trial_num'], bins=bins_scalar)
    arm2_rt['rt_bins'] = pd.cut(x=arm2_rt['trial_num'], bins=bins_scalar)
    arm3_rt['rt_bins'] = pd.cut(x=arm3_rt['trial_num'], bins=bins_scalar)
    i_series = range(1,mult_number)
    for scalar in i_series:
        d2_bins = d2_rt[(d2_rt['trial_num']> (bins * scalar - bins)) & (d2_rt['trial_num'] <= (bins * scalar))]
        arm1_bins = arm1_rt[(arm1_rt['trial_num']> (bins * scalar - bins)) & (arm1_rt['trial_num'] <= (bins * scalar))]
        arm2_bins = arm2_rt[(arm2_rt['trial_num']> (bins * scalar - bins)) & (arm2_rt['trial_num'] <= (bins * scalar))]
        arm3_bins = arm3_rt[(arm3_rt['trial_num']> (bins * scalar - bins)) & (arm3_rt['trial_num'] <= (bins * scalar))]
        trial_num_rep = bins                                              
        d2_mean = d2_bins.loc[d2_bins.feedback_code == 1,'rt'].mean()
        arm1_mean = arm1_bins.loc[arm1_bins.feedback_code == 1,'rt'].mean()
        arm2_mean = arm2_bins.loc[arm2_bins.feedback_code == 1,'rt'].mean()
        arm3_mean = arm3_bins.loc[arm3_bins.feedback_code == 1,'rt'].mean()
        d2_std = d2_bins.loc[d2_bins.feedback_code == 1,'rt'].std()
        arm1_std = arm1_bins.loc[arm1_bins.feedback_code == 1,'rt'].std()
        arm2_std = arm2_bins.loc[arm2_bins.feedback_code == 1,'rt'].std()
        arm3_std = arm3_bins.loc[arm3_bins.feedback_code == 1,'rt'].std()
        d2.append(d2_mean)
        arm21.append(arm1_mean)
        arm22.append(arm2_mean)
        arm23.append(arm3_mean)
        d2_err.append(d2_std)
        arm21_err.append(arm1_std)
        arm22_err.append(arm2_std)
        arm23_err.append(arm3_std)
    block_rt['trial_end'] = bins_scalar[1:]
    block_rt['d2_rt'] = d2
    block_rt['arm21_rt'] = arm21
    block_rt['arm22_rt'] = arm22
    block_rt['arm23_rt'] = arm23
    block_rt['d2_std'] = d2_err
    block_rt['arm21_std'] = arm21_err
    block_rt['arm22_std'] = arm22_err
    block_rt['arm23_std'] = arm23_err
    block_rt = block_rt.fillna(0)
    block_rt['trial_bin'] = pd.cut(x=block_rt['trial_end'], bins=bins_scalar)
    return block_rt
block_rt_sum(12,2,5)
```

```python
def graph_all_rt_1_group(bins,g_type):
  
    df1 = group_all_rt_1_Rmin(block = 1,bins = bins,mask = True,graph_type = g_type)
    df2 = group_all_rt_1_Rmin(block = 3,bins = bins,mask = True,graph_type = g_type)
    df3 = group_all_rt_1_Rmin(block = 2,bins = bins,mask = True,graph_type = g_type)
    df4 = group_all_rt_1_Rmin(block = 4,bins = bins,mask = True,graph_type = g_type)

    max_1 = df1['rt'].max()
    max_2 = df2['rt'].max()
    max_3 = df3['rt'].max()
    max_4 = df4['rt'].max()
    max_rt = max([max_1,max_2,max_3,max_4])
    min_1 = df1['rt'].min()
    min_2 = df2['rt'].min()
    min_3 = df3['rt'].min()
    min_4 = df4['rt'].min()
    min_rt = min([min_1,min_2,min_3,min_4])
    
    cat = "rt_bins"
    subcat = "pic_type"
    val = "rt"
    err = "std"
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 15))

    #block1
    u1 = df1[cat].unique()
    x1 = np.arange(len(u1))
    subx1 = df1[subcat].unique()
    offsets1 = (np.arange(len(subx1))-np.arange(len(subx1)).mean())/(len(subx1)+1.)
    width1 = np.diff(offsets1).mean()
    x1_label= list(df1.loc[df1.pic_type == 'decision1','rt_bins'])
    for i,gr in enumerate(subx1):
        dfg = df1[df1[subcat] == gr]
        axes[0,0].bar(x1+offsets1[i], dfg[val].values, width=width1, 
                label="{} {}".format(subcat, gr), yerr=dfg[err].values)
        axes[0,0].set_xticks(x1)
        axes[0,0].set_ylim(round(min_rt)-1,round(max_rt)+1)
        axes[0,0].set_yticks(np.arange(round(min_rt)-1,round(max_rt)+1,step=0.25))
        
        axes[0,0].set_xticklabels(x1_label,rotation = 45)
        axes[0,0].set_xlabel("trial_bins")
        axes[0,0].set_ylabel("rt average")
        axes[0,0].set_title( 'block_1, all rt, bins=' + str(bins))
        axes[0,0].legend(loc='upper right')

    #block3
    u2 = df2[cat].unique()
    x2 = np.arange(len(u2))
    subx2 = df2[subcat].unique()
    offsets2 = (np.arange(len(subx2))-np.arange(len(subx2)).mean())/(len(subx2)+1.)
    width2 = np.diff(offsets2).mean()
    x2_label= list(df2.loc[df2.pic_type == 'decision1','rt_bins'])
    for i,gr in enumerate(subx2):
        dfg = df2[df2[subcat] == gr]
        axes[0,1].bar(x2+offsets2[i], dfg[val].values, width=width2, 
                label="{} {}".format(subcat, gr), yerr=dfg[err].values)
        axes[0,1].set_xticks(x2)
        axes[0,1].set_ylim(round(min_rt)-1,round(max_rt)+1)
        axes[0,1].set_yticks(np.arange(round(min_rt)-1,round(max_rt)+1,step=0.25))
        
        axes[0,1].set_xticklabels(x2_label,rotation = 45)
        axes[0,1].set_xlabel("trial_bins")
        axes[0,1].set_ylabel("rt average")
        axes[0,1].set_title('block_3, all rt, bins=' + str(bins))
        axes[0,1].legend(loc='upper right')
        
    #block2
    u3 = df3[cat].unique()
    x3 = np.arange(len(u3))
    subx3 = df3[subcat].unique()
    offsets3 = (np.arange(len(subx3))-np.arange(len(subx3)).mean())/(len(subx3)+1.)
    width3 = np.diff(offsets3).mean()
    x3_label= list(df3.loc[df3.pic_type == 'decision1','rt_bins'])
    for i,gr in enumerate(subx3):
        dfg = df3[df3[subcat] == gr]
        axes[1,0].bar(x3+offsets3[i], dfg[val].values, width=width3, 
                label="{} {}".format(subcat, gr), yerr=dfg[err].values)
        axes[1,0].set_xticks(x3)
        
        axes[1,0].set_ylim(round(min_rt)-1,round(max_rt)+1)
        axes[1,0].set_yticks(np.arange(round(min_rt)-1,round(max_rt)+1,step=0.25))
        axes[1,0].set_xticklabels(x3_label,rotation = 45)
        axes[1,0].set_xlabel("trial_bins")
        axes[1,0].set_ylabel("rt average")
        axes[1,0].set_title('block_2, all rt, bins=' + str(bins))
        axes[1,0].legend(loc='upper right')
        
    #block4
    u4 = df4[cat].unique()
    x4 = np.arange(len(u4))
    subx4 = df4[subcat].unique()
    offsets4 = (np.arange(len(subx4))-np.arange(len(subx4)).mean())/(len(subx4)+1.)
    width4 = np.diff(offsets4).mean()
    x4_label= list(df4.loc[df4.pic_type == 'decision1','rt_bins'])
    for i,gr in enumerate(subx4):
        dfg = df4[df4[subcat] == gr]
        axes[1,1].bar(x4+offsets4[i], dfg[val].values, width=width4, 
                label="{} {}".format(subcat, gr), yerr=dfg[err].values)
        axes[1,1].set_xticks(x4)
        
        axes[1,1].set_ylim(round(min_rt)-1,round(max_rt)+1)
        axes[1,1].set_yticks(np.arange(round(min_rt)-1,round(max_rt)+1,step=0.25))
        axes[1,1].set_xticklabels(x4_label,rotation = 45)
        axes[1,1].set_xlabel("trial_bins")
        axes[1,1].set_ylabel("rt average")
        axes[1,1].set_title('block_4, all rt, bins=' + str(bins))
        axes[1,1].legend(loc='upper right')
   
    plt.savefig('group_all_RT_errbar_wMask' + str(3) + '_mean.png')
    plt.close()
graph_all_rt_1_group(bins = 10,g_type = "mean")
```
