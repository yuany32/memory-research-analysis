```python
#the list of training & testing fnames
fnames = sorted(glob.glob('*.csv'))
training_files = [fname for fname in fnames if re.match('(\d*)_main_exp_(\w*)_(\w*)_(\w*)_(\d*_\w*_\d*)_(\d*)_data.csv', fname)]
testing_files = [fname for fname in fnames if re.match('(\d*)_main_exp_(\w*)_(\w*)_(\w*)_(\d*_\w*_\d*)_(\d*).csv', fname)]
```

```python
#adding file_name colomn to each file
list_of_dfs = [pd.read_csv(fname) for fname in training_files]
for dataframe, fname in zip(list_of_dfs, training_files):
    dataframe['file_name'] = fname
```

```python
# Combine a list of dataframes, on top of each other
training = pd.concat(list_of_dfs, ignore_index=True)
```

```python
#merging block_order, block_order_bytype to each training file
info = pd.read_csv("fnames_info.csv")
training_data = pd.merge(training,
                 info[['file_name', 'block_order', 'block_order_bytype','subj_id']],
                 on ='file_name')
training_data = training_data.sort_values(by = ['subj_id','block_order'])
training_data = training_data.reset_index(drop = True)
training_data.head()
```

```python
#learning criterion for each subject each block
def trial_criteria (subj_id,block,criteria):
    
    # select specific subject and block data from trainig_data
    subj_acc = training_data.loc[training_data.subj_id == subj_id]
    block_data = subj_acc.loc[subj_acc.block_order == block]
    block_data = block_data.reset_index(drop = True)
    
    #create a new dataframe that has column of trial_num and percentage of correct feedback
    #for each trial
    block_criteria = pd.DataFrame(columns = ['trial_num','feedback_sum'])
    
    # Get the percentage of correct feedback for each trial
    feedback_sum = []
    for i in range (1,block_data['trial_num'].iloc[-1]+1): # <------- use trial_num as a indicator for iterating through rows
        trial_sum = (block_data.loc[block_data['trial_num'] == i, 'feedback_code'].sum())/12
        feedback_sum.append(trial_sum)
        i += 1 
    block_criteria['trial_num'] = np.arange(len(feedback_sum)) + 1
    block_criteria['feedback_sum'] = feedback_sum

    #count the correct number of trials 
    correct = 0 
    percent = 0
    correct_count = []
    correct_percent = []
    for i in range (0,len(block_criteria)-(criteria-1)):
        iterate_trial = (block_criteria.iloc[i:(i+criteria)])['feedback_sum']
        for iterate_rows in iterate_trial:
            if iterate_rows == 1:
                correct += 1
                percent = correct / criteria
        correct_count.append(correct)
        correct_percent.append(percent)
        correct = 0
        percent = 0
    
    #create a new dataframe that has trial_num, overall_percentage
    overall_criteria = pd.DataFrame(columns = ['trial_criteria','overall_data'])
    overall_criteria['trial_criteria'] = np.arange(len(block_criteria['trial_num'])-(criteria-1)) + 1
    overall_criteria['overall_data'] = correct_percent
    block_data.loc[block_data.trial_target == "bolt",'trial_target'] = "lion"
    block_data.loc[block_data.trial_target == "arc",'trial_target'] = "beaver"
    overall_criteria['trial_target'] = block_data['trial_target'][0] 
    overall_criteria['count'] = correct_count
    overall_criteria['trial_sq'] = overall_criteria['trial_criteria']**2

    return overall_criteria
```

```python
def trial_criteria_graph(subj_id,criteria):
    fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 5))
    for j in range(1,5):
        overall_criteria = trial_criteria (subj_id,j,criteria)

        #get regression line for learning criteria
        #change sigmoid    
        formula_criteria = 'overall_data ~ trial_sq + trial_criteria'
        mod = ols(formula=formula_criteria, data=overall_criteria)
        results = mod.fit()
        #print(results.summary())
        predicted = results.predict()
        #print(predicted)

        # scatter plot of learning criteria
        target = overall_criteria['trial_target'][0]
        if target == "arc":
            target = "beaver"
        elif target == "bolt":
            target = "lion"
        axes[j-1].set_title('trial accuracy (criteria = ' + str(criteria) + ')',loc = 'center')
        axes[j-1].scatter(overall_criteria['trial_criteria'],overall_criteria['overall_data'],label = 'subj_' + str(subj_id) \
                + ', block ' + str(j) + ', ' + 'criteria = ' + str(criteria) + ', '+ target )
        axes[j-1].plot(overall_criteria['trial_criteria'], predicted, linewidth = 3) # <------------- Add predicted values
        axes[j-1].set_xlabel('trial bins') # <--------------- Set x axis label
        axes[j-1].legend(loc = "lower right")
        plt.yticks(np.arange(0, 1.2, step=0.2))
        axes[j-1].set_ylim(0, 1)
        axes[j-1].set_xlim(0, len(overall_criteria['trial_criteria']))
        axes[j-1].set_ylabel('Accuracy (percentage)') # <--------- Set y axis label
        j += 1
       
    #save subplots for each subject
    #plt.savefig('subj'+str(subj_id)+ '_criteria' + str(criteria) + '_trial_criteria.png')
    #plt.close() 
```

```python
def trial_response_time(subj_id,block,criteria):
     # select specific subject and block data from trainig_data
    subj_acc = training_data.loc[training_data.subj_id == subj_id]
    block_data = subj_acc.loc[subj_acc.block_order == block]
    block_data = block_data.reset_index(drop = True)
    
    #create a new dataframe that has column of trial_num and percentage of correct feedback
    #for each trial
    block_rt = pd.DataFrame(columns = ['trial_num','rt'])
    
    # Get the percentage of correct feedback for each trial
    rt_sum = []
    for i in range (1,block_data['trial_num'].iloc[-1]+1): # <------- use trial_num as a indicator for iterating through rows
        rt = (block_data.loc[block_data['trial_num'] == i, 'rt'].mean())
        rt_sum.append(rt)
        i += 1 
    block_rt['trial_num'] = np.arange(len(rt_sum)) + 1
    block_rt['rt'] = rt_sum
    
    #sum rt for each criteria
    rt_criteria = []
    for i in range (0,len(block_rt)-(criteria-1)):
        iterate_trial = (block_rt.iloc[i:(i+criteria)])['rt']
        rt = iterate_trial.sum()
        rt_criteria.append(rt)
        
    #create a new dataframe that has trial_num, overall_percentage
    overall_rt = pd.DataFrame(columns = ['trial_criteria','rt_sum'])
    overall_rt['trial_criteria'] = np.arange(len(block_rt['trial_num'])-(criteria-1)) + 1    
    overall_rt['rt_sum'] = rt_criteria
    return overall_rt
#trial_response_time(subj_id = 5,block=1,criteria = 5)
```

```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit 
import scipy.optimize as optimize
def trial_rt_criteria_graph(subj_id,criteria):
    fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))
    for j in range(1,5):
        overall_criteria = trial_criteria (subj_id = subj_id,block = j,criteria = criteria)
        rt_criteria = trial_response_time(subj_id = subj_id,block = j,criteria = criteria)
        
        #sigmoid for all trials criteria
        formula_criteria = 'overall_data ~ trial_sq + trial_criteria'
        mod = ols(formula=formula_criteria, data=overall_criteria)
        results = mod.fit()
        predicted = results.predict()
    
        # log formula
        def func(t, a, b):
            return a + b * np.log(t)
        popt, pcov = optimize.curve_fit(func,rt_criteria['trial_criteria'] , rt_criteria['rt_sum'], maxfev=5000)
        t = rt_criteria['trial_criteria']
        
        target = overall_criteria['trial_target'][0]
       
        #set up x_lim & y_lim
        #if (j == 1) | (j == 2):
        max_rt1 = max(trial_response_time(subj_id = subj_id,block = 1,criteria = criteria)['rt_sum'])
        max_rt2 = max(trial_response_time(subj_id = subj_id,block = 2,criteria = criteria)['rt_sum'])
        if max_rt1 > max_rt2:
            max_rt = max_rt1
        else:
            max_rt = max_rt2
                
        axes[0,j-1].set_title('trial accuracy (criteria = ' + str(criteria) + ')',loc = 'center',fontsize=14)
        axes[0,j-1].scatter(overall_criteria['trial_criteria'],overall_criteria['overall_data'],label = 'subj_' + str(subj_id) \
                + ', block ' + str(j) + ', ' + 'criteria = ' + str(criteria) + ', '+ target )
        axes[0,j-1].plot(overall_criteria['trial_criteria'], predicted, linewidth = 3) # <------------- Add predicted values
        
        # scatter plots settings
        axes[0,j-1].set_xlabel('trial bins',fontsize=14) # <--------------- Set x axis label
        axes[0,j-1].legend(loc = "lower right")
        plt.yticks(np.arange(0, 1.2, step=0.2))
        axes[0,j-1].set_ylim(0, 1)
        axes[0,j-1].set_xlim(0,len(overall_criteria['trial_criteria']))
        axes[0,j-1].set_ylabel('Accuracy (percentage)',fontsize=14) # <--------- Set y axis label
        
        plt.tight_layout()
        axes[1,j-1].set_title('trial rt (criteria = ' + str(criteria) + ')',loc = 'center',fontsize=14)
        axes[1,j-1].scatter(rt_criteria['trial_criteria'],rt_criteria['rt_sum'],label = 'subj_' + str(subj_id) \
                + ', block ' + str(j) + ', ' + 'criteria = ' + str(criteria) + ', ' + target + ', rt' ,color = 'green')
        axes[1,j-1].plot(rt_criteria['trial_criteria'], func(rt_criteria['trial_criteria'],*popt),label = '',
                       linewidth = 3,color = 'green') # <------------- Add predicted values
        
        # scatter plot settings
        axes[1,j-1].set_xlabel('trial bins',fontsize=14) # <--------------- Set x axis label
        axes[1,j-1].legend(loc = "lower right")
        axes[1,j-1].set_yticks(np.arange(0,(max_rt + 1),step=(max_rt / 5)))
        axes[1,j-1].set_ylim(0, max_rt)
        axes[1,j-1].set_xlim(0,len(overall_criteria['trial_criteria']))
        axes[1,j-1].set_ylabel('response time (average)',fontsize=14) # <--------- Set y axis label
        plt.rc('xtick',labelsize= 14)
        plt.rc('ytick',labelsize= 14)
        j += 1
       
    
    #save subplots for each subject
    plt.savefig('subj'+str(subj_id)+ '_criteria' + str(criteria) + '_trial_rt.png')
    plt.close() 
```

```python
#get the max percent of achievement for all subjects
def max_achieved(criteria,subj_bins):
    # create a new df that has max acheived for each block per subject 
    block_1 = []
    block_2 = []
    block_3 = []
    block_4 = []
    max_trials_1 = []
    max_trials_2 = []
    max_trials_3 = []
    max_trials_4 = []
    for i in range(1,5):
        for j in range(1,(subj_bins + 1)):
            correct_percent = trial_criteria (subj_id = j,block = i,criteria = criteria)['overall_data']
            trial = trial_criteria (subj_id = j,block = i,criteria = criteria)['trial_criteria']
            block_max = correct_percent.max()
            trial_num = trial[correct_percent.idxmax()]
            if correct_percent.nunique() == 1:
                trial_num = 1
            if i == 1:
                block_1.append(block_max)
                max_trials_1.append(trial_num) 
            elif i == 2:
                block_2.append(block_max)
                max_trials_2.append(trial_num) 
            elif i == 3:
                block_3.append(block_max)
                max_trials_3.append(trial_num) 
            else:
                block_4.append(block_max)
                max_trials_4.append(trial_num) 
        
    overall_max = pd.DataFrame(columns = ['subj_id','block_1','block_2',
                                          'block_3','block_4','max_trials_1','max_trials_2',
                                         'max_trials_3','max_trials_4'])
    overall_max['subj_id'] = np.arange(1,(subj_bins + 1))
    overall_max['block_1'] = block_1
    overall_max['block_2'] = block_2
    overall_max['block_3'] = block_3
    overall_max['block_4'] = block_4
    
    #columns that have the trial numbers when each subject achieve max performance
    overall_max['max_trials_1'] = max_trials_1
    overall_max['max_trials_2'] = max_trials_2
    overall_max['max_trials_3'] = max_trials_3
    overall_max['max_trials_4'] = max_trials_4
    return overall_max
            #print(correct_percent)
```

```python
# histogram for max performance achieved for all subjects
binBoundaries = np.linspace(0,1,5)
def max_count_hist(criteria,binBoundaries):
    fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 5))
    # Defining custom 'xlim' and 'ylim' values.
    xlim = (0, 1)
    ylim = (0, 30)

    # Setting the values for all axes.
    plt.setp(axes, xlim = xlim, ylim=ylim)
    max_count = max_achieved(criteria,subj_bins = 30)
    axes[0].hist(np.asarray(max_count['block_1'], dtype='float'),histtype='bar',bins = binBoundaries, orientation='vertical',label = 'block_1')
    axes[0].set_title('block_1 Max % achieved, criteria ' + str(criteria))
    axes[1].hist(np.asarray(max_count['block_2'], dtype='float'),histtype='bar',bins = binBoundaries, orientation='vertical',label = 'block_2')
    axes[1].set_title('block_2 Max % achieved, criteria ' + str(criteria))
    axes[2].hist(np.asarray(max_count['block_3'], dtype='float'),bins = binBoundaries, orientation='vertical',label = 'block_3')
    axes[2].set_title('block_3 Max % achieved, criteria ' + str(criteria))
    axes[3].hist(np.asarray(max_count['block_4'], dtype='float'),bins = binBoundaries, orientation='vertical',label = 'block_4')
    axes[3].set_title('block_4 Max % achieved, criteria ' + str(criteria))
    plt.gca().set(ylabel= 'Frequency',xlabel = 'max percentage(%)')
    plt.savefig('criteria' + str(criteria) + '_max% performance.png')
    plt.close() 
    return fig,axes
#max_count_hist(criteria = 15,binBoundaries = binBoundaries)
```

```python
def failed_threshold(criteria):
    max_count = max_achieved(criteria = criteria, subj_bins = 30)
    threshold = max_count.mode()
    threshold = threshold['block_1'][0]
    #threshold = 
    print('==============================subject failed max% performance threshold, criteria = ' + str(criteria) + '============================')
    print('threshold = ' +str(threshold))
    
    block_1 = pd.DataFrame({'subj_id':max_count.loc[(max_count['block_1'] < threshold)]['subj_id']})#.reset_index(drop = True)
    block_1_target = []
    for subj in block_1['subj_id']:
        block_1_type = trial_criteria (subj_id = subj ,block = 1,
                                      criteria = criteria)['trial_target'][0]
        block_1_target.append(block_1_type)
    block_1['block_1_target'] = block_1_target
    print('<------------block_1, ' + 'creteria=' + str(criteria) + ', subject below threshold--------------->')
    print(block_1.reset_index(drop = True))
    
    block_2 = pd.DataFrame({'subj_id':max_count.loc[(max_count['block_2'] < threshold)]['subj_id']})#.reset_index(drop = True)
    block_2_target = []
    for subj in block_2['subj_id']:
        block_2_type = trial_criteria (subj_id = subj ,block = 2,
                                      criteria = criteria)['trial_target'][0]
        block_2_target.append(block_2_type)
    block_2['block_2_target'] = block_2_target
    print('<------------block_2, ' + 'creteria=' + str(criteria) + ', subject below threshold--------------->')
    print(block_2.reset_index(drop = True))
    print()
    
    block_3 = pd.DataFrame({'subj_id':max_count.loc[(max_count['block_3'] < threshold)]['subj_id']})#.reset_index(drop = True)
    block_3_target = []
    for subj in block_3['subj_id']:
        block_3_type = trial_criteria (subj_id = subj, block = 3, criteria = criteria)['trial_target'][0]
        block_3_target.append(block_3_type)
    block_3['block_3_target'] = block_3_target
    print('<------------block_3, ' + 'creteria=' + str(criteria) + ', subject below threshold--------------->')
    print(block_3.reset_index(drop = True))
    print()
    
    block_4 = pd.DataFrame({'subj_id':max_count.loc[(max_count['block_4'] < threshold)]['subj_id']})#.reset_index(drop = True)
    block_4_target = []
    for subj in block_4['subj_id']:
        block_4_type = trial_criteria(subj_id = subj, block = 4,criteria = criteria)['trial_target'][0]
        block_4_target.append(block_4_type)
    block_4['block_4_target'] = block_4_target
    print('<------------block_4, ' + 'creteria=' + str(criteria) + ', subject below threshold--------------->')
    print(block_4.reset_index(drop = True)) 
    print()
    
    print("<------------more than 1 block below threshold------------>")
    perf_below_threshold = pd.DataFrame(columns=['criteria','threshold','subj_id'])
    subj_id = np.arange(1,31)
    perf_below_threshold['subj_id'] = subj_id
    perf_below_threshold['criteria'] = [criteria]*len(subj_id)
    perf_below_threshold['threshold'] = [threshold]*len(subj_id)
    #perf_below_threshold['block_1']= block_1
    #perf_below_threshold['block_2']= block_2
    #perf_below_threshold['block_3']= block_3
    #perf_below_threshold['block_4']= block_4
    perf_below_threshold = pd.merge(perf_below_threshold,
                     block_1[['subj_id', 'block_1_target']],
                     on ='subj_id',how = 'left')    
    perf_below_threshold = pd.merge(perf_below_threshold,
                     block_2[['subj_id', 'block_2_target']],
                     on ='subj_id',how = 'left')    
    perf_below_threshold = pd.merge(perf_below_threshold,
                     block_3[['subj_id', 'block_3_target']],
                     on ='subj_id',how = 'left') 
    perf_below_threshold = pd.merge(perf_below_threshold,
                     block_4[['subj_id', 'block_4_target']],
                     on ='subj_id',how = 'left') 
    
    return perf_below_threshold
```

```python
def all_failed_threshold (criteria):
    perf_below_threshold = failed_threshold(criteria = criteria)
    perf_below_threshold = pd.merge(perf_below_threshold,
                     max_achieved(criteria,30)[['subj_id','max_trials_1',
                      'max_trials_2','max_trials_3',
                      'max_trials_4']],
                     on = 'subj_id',how = 'left')
    perf_below_threshold = pd.merge(perf_below_threshold,
                     max_achieved(criteria,30)[['subj_id','block_1',
                      'block_2','block_3',
                      'block_4']],
                     on = 'subj_id',how = 'left')
    perf_below_threshold = perf_below_threshold.rename(columns={"block_1": "max_perf_1", "block_2": "max_perf_2",
                                         "block_3": "max_perf_3", "block_4": "max_perf_4", 
                                         "max_trials_1": "reach_max_1","max_trials_2": "reach_max_2",
                                         "max_trials_3": "reach_max_3","max_trials_4": "reach_max_4"})
    perf_below_threshold = perf_below_threshold[['criteria','threshold','subj_id','max_perf_1','max_perf_2',
                                              'max_perf_3','max_perf_4','block_1_target','block_2_target'
                                              ,'block_3_target','block_4_target','reach_max_1','reach_max_2',
                                              'reach_max_3','reach_max_4']]
    #pd.concat([perf_below_threshold,block2_below], ignore_index=True, axis=1)
    perf_below_threshold = perf_below_threshold.reset_index(drop = True)
    
    return perf_below_threshold
```

```python
def group_criteria(blocks,criterias,target,num):
    mean = []
    std= []

    group_accuracy = pd.DataFrame(columns = ['trial_num','mean','std'])
    accuracy_df = []
    for subj in range(1,31):
        trial_df =  trial_criteria (subj_id = subj,block = blocks,criteria = criterias)
        if target == "lion":
            df =trial_df.loc[trial_df.trial_target == "lion"]
            df['subj_id'] = [str(subj)] * len(df)
        elif target == "beaver":
            df =trial_df.loc[trial_df.trial_target == "beaver"]
            df['subj_id'] = [str(subj)] * len(df)
        accuracy_df.append(df)
    accuracy_df = pd.concat(accuracy_df)
    trial_num = accuracy_df.loc[accuracy_df.subj_id == str(num),'trial_criteria']
    #trial_num = accuracy_df.loc[accuracy_df.subj_id == "2",'trial_criteria']
    for trial in trial_num:
        percentage = accuracy_df.loc[accuracy_df.trial_criteria == trial,"overall_data"].mean()
        count = len(accuracy_df.loc[accuracy_df.trial_criteria == trial])
        stdev = accuracy_df.loc[accuracy_df.trial_criteria == trial,"overall_data"].std() / np.sqrt(count)
        mean.append(percentage)
        std.append(stdev)
    group_accuracy['trial_num'] = trial_num
    group_accuracy['mean'] = mean
    group_accuracy['std'] = std
    group_accuracy['trial_target'] = [target] * len(group_accuracy)
    return group_accuracy
```

```python
group_criteria(blocks = 3,criterias = 10,target = "lion", num = 1)
```

```python
def graph_target(criterias):
    df1= group_criteria(1,criterias,"lion",1)
    df2 = group_criteria(2,criterias,"beaver",1)
    df3 = group_criteria(3,criterias,"lion",1)
    df4 = group_criteria(4,criterias,"beaver",1)
    
    df5= group_criteria(1,criterias,"beaver",2)
    df6 = group_criteria(2,criterias,"lion",2)
    df7 = group_criteria(3,criterias,"beaver",2)
    df8 = group_criteria(4,criterias,"lion",2)
    max_1 = df1['mean'].max()
    max_2 = df2['mean'].max()
    max_3 = df3['mean'].max()
    max_4 = df4['mean'].max()
    max_all = max([max_1,max_2,max_3,max_4])
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize = (14,14))
    ax = axes[0,0]
    ax.errorbar(df1['trial_num'], df1['mean'], yerr=df1['std'],fmt='o',color = 'green',label = "block1, lion")
    ax.errorbar(df2['trial_num'], df2['mean'], yerr=df2['std'],fmt='o',color = 'blue',label = "block2, beaver")
    ax.legend(loc='lower right')
    ax.set_yticks(np.arange(0,1.2,step = 0.25))
    ax.set_xticks(np.arange(0,len(df1)+1,step=8))
    ax.set_title("group performance, block_1-2, criteria = " + str(criterias))
    ax.set_xlabel('trial_num')
    ax.set_ylabel('performance(percentage)')
    ax = axes[0,1]
    ax.errorbar(df3['trial_num'], df3['mean'], yerr=df3['std'],color = 'green',fmt='o', label = "block3, lion ")
    ax.errorbar(df4['trial_num'], df4['mean'], yerr=df4['std'],color = 'blue',fmt='o', label = "block4, beaver ")
    ax.legend(loc='lower right')
    ax.set_yticks(np.arange(0,1.2,step = 0.25))
    ax.set_xticks(np.arange(0,len(df3)+1,step=8))
    ax.set_title("group performance, block_3-4, criteria = " + str(criterias))
    ax.set_xlabel('trial_num')
    ax.set_ylabel('performance(percentage)')
    ax = axes[1,0]
    ax.errorbar(df5['trial_num'], df5['mean'], yerr=df5['std'],color = 'green',fmt='o',label = "block1, beaver")
    ax.errorbar(df6['trial_num'], df6['mean'], yerr=df6['std'],color = 'blue',fmt='o',label = "block2, lion" )
    ax.legend(loc='lower right')
    ax.set_yticks(np.arange(0,1.2,step = 0.25))
    ax.set_xticks(np.arange(0,len(df5)+1,step=8))
    ax.set_title("group performance, block_1-2, criteria = " + str(criterias))
    ax.set_xlabel('trial_num')
    ax.set_ylabel('performance(percentage)')
    ax = axes[1,1]
    ax.errorbar(df7['trial_num'], df7['mean'], yerr=df7['std'],color = 'green',fmt='o',label = "block3, beaver")
    ax.errorbar(df8['trial_num'], df8['mean'], yerr=df8['std'],color = 'blue',fmt='o', label = "block4, lion")
    ax.legend(loc='lower right')
    ax.set_yticks(np.arange(0,1.2,step = 0.25))
    ax.set_xticks(np.arange(0,len(df7)+1,step=8))
    ax.set_title("group performance, block_3-4, criteria = " + str(criterias))
    ax.set_xlabel('trial_num')
    ax.set_ylabel('performance(percentage)')
    plt.savefig('group_accuracy_by_trial_criteria_target' + str(criterias)+ '.png')
graph_target(criterias = 10)
```
